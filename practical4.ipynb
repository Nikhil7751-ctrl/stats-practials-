1 Calculate the percentage of visible minonties are tenure professors. Will you say that tenure status differed if teacher was a visible minority? 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

faculty_data = {
    'ethnicity': np.random.choice(['Minority', 'Non-Minority'], 60, p=[0.35, 0.65]),
    'permanent_position': np.random.choice(['Yes', 'No'], 60, p=[0.55, 0.45]),
    'years_of_experience': np.random.randint(25, 62, 60),
    'sex': np.random.choice(['Male', 'Female'], 60),
    'performance_score': np.round(np.random.normal(4.2, 0.4, 60), 2)
}
```

```python
df = pd.DataFrame(faculty_data)
permanent_minority = df[(df['permanent_position'] == 'Yes') & (df['ethnicity'] == 'Minority')]
total_minority = df[df['ethnicity'] == 'Minority']

minority_ratio = (len(permanent_minority) / len(total_minority)) * 100
print(f"⿡ Percentage of minority group with permanent positions: {minority_ratio:.2f}%")

if minority_ratio < 50:
    print("Minority staff have a lower rate of permanent positions.")
else:
    print("Permanent positions are fairly distributed across groups.")
```

2. Does average age differ by tenure? Produce the means and standard deviations for both fenured and untenured professors

```python

permanent = df[df['permanent_position'] == 'Yes']['years_of_experience']
temporary = df[df['permanent_position'] == 'No']['years_of_experience']

print("\n⿢ Average years of experience by job status:")
print(f"Permanent - Mean: {permanent.mean():.2f}, Std Dev: {permanent.std():.2f}")
print(f"Temporary - Mean: {temporary.mean():.2f}, Std Dev: {temporary.std():.2f}")

plt.figure(figsize=(6, 4))
plt.boxplot([permanent, temporary], labels=['Permanent', 'Temporary'], patch_artist=True)
plt.title("Experience Distribution by Job Status")
plt.ylabel("Years of Experience")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

```

```python
print("\n⿤ Difference between plt.bar() and plt.barh():")
print("• plt.bar() → vertical bars (X-axis categories, Y-axis values)")
print("• plt.barh() → horizontal bars (useful for long category names or better readability)")

gender_counts = df['sex'].value_counts()

plt.figure(figsize=(5, 4))
plt.bar(gender_counts.index, gender_counts.values, color=['teal', 'orange'])
plt.title("Gender Distribution (Vertical Bar Chart)")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(5, 4))
plt.barh(gender_counts.index, gender_counts.values, color=['teal', 'orange'])
plt.title("Gender Distribution (Horizontal Bar Chart)")
plt.xlabel("Count")
plt.show()
```
